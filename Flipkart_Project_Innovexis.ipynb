## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

````python
# Three Hypothetical Statements for Statistical Testing

print("=" * 80)
print("HYPOTHETICAL STATEMENTS FOR TESTING")
print("=" * 80)

print("\n1. HYPOTHESIS 1: Channel Effectiveness")
print("-" * 80)
print("Statement: Different communication channels (chat, call, email) have")
print("significantly different average CSAT scores.")
print("\nNull Hypothesis (H0): All channels have the same average CSAT score")
print("Alternative Hypothesis (H1): At least one channel has a different average CSAT score")
print("Test: ANOVA (Analysis of Variance)")

print("\n\n2. HYPOTHESIS 2: Agent Experience Impact")
print("-" * 80)
print("Statement: Agents with longer tenure (experience) provide significantly")
print("better customer satisfaction than those with shorter tenure.")
print("\nNull Hypothesis (H0): Tenure does not affect CSAT scores")
print("Alternative Hypothesis (H1): Tenure has a significant effect on CSAT scores")
print("Test: ANOVA (Analysis of Variance)")

print("\n\n3. HYPOTHESIS 3: Handling Time Correlation")
print("-" * 80)
print("Statement: There is a significant correlation between connected handling")
print("time and CSAT score (i.e., faster resolution leads to higher satisfaction).")
print("\nNull Hypothesis (H0): No correlation between handling time and CSAT score")
print("Alternative Hypothesis (H1): Significant correlation exists between handling time and CSAT")
print("Test: Pearson Correlation Test")

print("\n" + "=" * 80)
print("Significance Level: α = 0.05")
print("=" * 80)
````

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

HYPOTHESIS 1: Channel Effectiveness

Research Question: Do different communication channels (chat, call, email, etc.) have significantly different average CSAT scores?

Null Hypothesis (H₀): There is no significant difference in the average CSAT scores across different communication channels. All channels have equal mean CSAT scores.

Alternative Hypothesis (H₁): There is a significant difference in the average CSAT scores across different communication channels. At least one channel has a different mean CSAT score compared to others.

Statistical Test: One-Way ANOVA (Analysis of Variance)

Significance Level: α = 0.05

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Hypothesis Test 1: Channel Effectiveness
# Testing if different channels have different CSAT scores

from scipy.stats import f_oneway

print("=" * 80)
print("HYPOTHESIS TEST 1: CHANNEL EFFECTIVENESS")
print("=" * 80)

# Separate CSAT scores by channel
channels = df['channel_name'].unique()
channel_groups = [df[df['channel_name'] == channel]['CSAT Score'].dropna() for channel in channels]

# Perform One-Way ANOVA
f_statistic, p_value = f_oneway(*channel_groups)

# Display results
print(f"\nTest: One-Way ANOVA")
print(f"F-Statistic: {f_statistic:.4f}")
print(f"P-Value: {p_value:.6f}")
print(f"Significance Level: α = 0.05")

print("\n" + "-" * 80)
print("CONCLUSION:")
print("-" * 80)
if p_value < 0.05:
    print("✓ REJECT the Null Hypothesis (H₀)")
    print("✓ There IS a significant difference in CSAT scores across channels")
    print(f"  (p-value = {p_value:.6f} < 0.05)")
else:
    print("✗ FAIL TO REJECT the Null Hypothesis (H₀)")
    print("✗ There is NO significant difference in CSAT scores across channels")
    print(f"  (p-value = {p_value:.6f} >= 0.05)")

# Show channel-wise statistics
print("\n" + "-" * 80)
print("Channel-wise CSAT Statistics:")
print("-" * 80)
print(df.groupby('channel_name')['CSAT Score'].agg(['mean', 'std', 'count']))
````

##### Which statistical test have you done to obtain P-Value?

I used **One-Way ANOVA (Analysis of Variance)** to obtain the P-Value.

**Why ANOVA?**
- We're comparing the **means** of CSAT scores across **multiple groups** (different channels like chat, call, email, etc.)
- ANOVA is the appropriate test when comparing means of **3 or more independent groups**
- It tests whether at least one group mean is significantly different from the others

For this hypothesis about channel effectiveness with multiple channels, **One-Way ANOVA** is the correct choice!

##### Why did you choose the specific statistical test?

Reasons for choosing One-Way ANOVA:

Multiple Groups to Compare: We have multiple channels (chat, call, email, etc.) - more than 2 groups. ANOVA is designed to compare means across 3+ groups simultaneously.

One Independent Variable: We have one categorical independent variable (channel_name) and we want to see its effect on one continuous dependent variable (CSAT Score).

Comparing Means: We want to test if the average CSAT scores differ across channels, and ANOVA specifically tests differences in means.

Efficient Testing: Instead of doing multiple t-tests (chat vs call, chat vs email, call vs email, etc.), ANOVA does it all in one test, which reduces the chance of Type I error (false positives).

Standard Practice: When the research question is "Does variable X differ across groups A, B, C?", ANOVA is the gold standard statistical test.

Continuous Dependent Variable: CSAT Score is numerical/continuous, and we're comparing the mean CSAT across different tenure groups.

Ordinal Nature with Multiple Levels: Even though tenure has an order (less experience → more experience), with multiple levels, ANOVA is still appropriate to test if ANY differences exist.

Statistical Efficiency: One test for all groups instead of multiple pairwise t-tests, reducing Type I error risk.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

HYPOTHESIS 3: Handling Time Correlation
Research Question: Is there a significant correlation between connected handling time and CSAT score? Does faster resolution lead to higher customer satisfaction?

Null Hypothesis (H₀): There is no significant correlation between connected handling time and CSAT score. Handling time does not affect customer satisfaction (ρ = 0).

Alternative Hypothesis (H₁): There is a significant correlation between connected handling time and CSAT score. Handling time does affect customer satisfaction (ρ ≠ 0).

Statistical Test: Pearson Correlation Test (or Spearman if data is not normally distributed)

Significance Level: α = 0.05

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Hypothesis Test 3: Handling Time Correlation
# Testing if there's a correlation between handling time and CSAT scores

from scipy.stats import pearsonr

print("=" * 80)
print("HYPOTHESIS TEST 3: HANDLING TIME CORRELATION")
print("=" * 80)

# Remove missing values
data_clean = df[['connected_handling_time', 'CSAT Score']].dropna()

# Perform Pearson Correlation Test
correlation_coef, p_value = pearsonr(data_clean['connected_handling_time'], data_clean['CSAT Score'])

# Display results
print(f"\nTest: Pearson Correlation Test")
print(f"Correlation Coefficient (r): {correlation_coef:.4f}")
print(f"P-Value: {p_value:.6f}")
print(f"Significance Level: α = 0.05")

print("\n" + "-" * 80)
print("CORRELATION INTERPRETATION:")
print("-" * 80)
if abs(correlation_coef) < 0.3:
    strength = "Weak"
elif abs(correlation_coef) < 0.7:
    strength = "Moderate"
else:
    strength = "Strong"

direction = "Positive" if correlation_coef > 0 else "Negative"
print(f"{strength} {direction} Correlation (r = {correlation_coef:.4f})")

print("\n" + "-" * 80)
print("CONCLUSION:")
print("-" * 80)
if p_value < 0.05:
    print("✓ REJECT the Null Hypothesis (H₀)")
    print("✓ There IS a significant correlation between handling time and CSAT")
    print(f"  (p-value = {p_value:.6f} < 0.05)")
    if correlation_coef < 0:
        print("✓ Longer handling time is associated with LOWER satisfaction")
    else:
        print("✓ Longer handling time is associated with HIGHER satisfaction")
else:
    print("✗ FAIL TO REJECT the Null Hypothesis (H₀)")
    print("✗ There is NO significant correlation between handling time and CSAT")
    print(f"  (p-value = {p_value:.6f} >= 0.05)")

print("\n" + "-" * 80)
print(f"Sample Size: {len(data_clean)} records")
print("-" * 80)
````

I used Pearson Correlation Test to obtain the P-Value.

Connected handling time: continuous (minutes/seconds)
CSAT Score: numerical (1-5 scale)
Correlation tests are specifically designed for two continuous variables


Testing Relationship, Not Group Differences:

We're not comparing groups (like in Hypotheses 1 & 2)
We're testing if two variables move together (as one changes, does the other change?)


Research Question is About Association:

"Is there a relationship/correlation?" → Use correlation test
"Are groups different?" → Use ANOVA/t-test


Linear Relationship Expected:

Pearson specifically measures linear correlation
We expect: as handling time increases/decreases, CSAT might consistently increase/decrease


Provides Both Strength and Significance:

Correlation coefficient (r): tells us HOW strong the relationship is (-1 to +1)
P-value: tells us if the relationship is statistically significant


Simple and Interpretable:

Easy to understand: negative r = inverse relationship, positive r = direct relationship

Given that numerical columns like Item_price can often have skewed distributions or outliers (as seen in the earlier df.describe() output), using the median helps to preserve the distribution of the data without being disproportionately affected by extreme values.

Categorical Columns (e.g., Customer Remarks, Order_id, order_date_time, Customer_City, Product_category): I filled missing values with the mode (most frequent value) of each respective column.

Why Mode? For categorical data, the mode is the most appropriate central tendency measure. Filling with the mode ensures that the imputed values are valid categories already present in the data and that the overall distribution of the categorical variable is minimally disturbed, especially when a single category dominates. Columns like Customer_City and Product_category are prime examples where mode imputation is suitable.

Preparation for Tree-based Models: For many tree-based machine learning algorithms (like Decision Trees, Random Forests, Gradient Boosting), Label Encoding often works well because these models can interpret numerical order and split data effectively even with arbitrarily assigned labels, unlike linear models which might misinterpret the numerical differences.

##### Which text normalization technique have you used and why?

I have used Lemmatization as the text normalization technique.

Why Lemmatization?

Reduction to Base Form: Lemmatization reduces words to their meaningful base form, or 'lemma' (e.g., 'running' becomes 'run', 'better' becomes 'good'). This helps standardize words and treats different grammatical forms of a word as a single item.

Linguistic Context: Unlike stemming (which simply chops off suffixes and can result in non-words), lemmatization uses vocabulary and morphological analysis to return valid words. This is crucial for maintaining the semantic integrity of the text, which is important for understanding customer remarks.

Improved Accuracy: By normalizing words, it helps machine learning models to avoid treating different forms of the same word as distinct features. This reduces the feature space, mitigates sparsity, and can improve the accuracy and generalization capability of models that rely on word frequencies or embeddings.

Better Interpretation: The resulting lemmas are actual dictionary words, making the preprocessed text more human-readable and easier to interpret, which is beneficial for qualitative analysis of customer feedback

##### Which text vectorization technique have you used and why?

I have used TF-IDF (Term Frequency-Inverse Document Frequency) for text vectorization.

Why TF-IDF?

Captures Term Importance: TF-IDF reflects how important a word is to a document in a collection or corpus. It increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus, which helps to adjust for the fact that some words appear more frequently in general.

Handles Stopwords (Implicitly): Highly frequent words across many documents (like 'the', 'a', 'is' if not already removed as stopwords) will have a low inverse document frequency, thereby reducing their weight and importance. This helps in focusing on more significant terms.

Widely Used and Effective: TF-IDF is a robust, interpretable, and widely used technique for converting textual data into a numerical format suitable for machine learning algorithms, especially in tasks like classification.

Feature Representation: It creates a numerical representation (vectors) where each dimension corresponds to a word in the vocabulary and the value indicates the TF-IDF score, allowing models to process textual information.

3.  **Good Performance with Fewer Resources**: RandomizedSearchCV often finds good or near-optimal solutions in a fraction of the time required by GridSearchCV, particularly when many hyperparameters don't have a significant impact on performance.

4.  **Early Exploration**: By sampling, it allows for quicker exploration of the hyperparameter space, helping to identify promising regions without exhaustive computation. This is particularly useful in the initial stages of model development.

3. Explain each evaluation metric's indication towards business and the business impact pf the ML model used.

Here's an explanation of each evaluation metric's indication towards business and the business impact of the ML models used, especially in the context of predicting CSAT scores for Flipkart:

### 1. Accuracy

*   **Indication:** Accuracy measures the proportion of total predictions that were correct. In the context of multi-class CSAT prediction, it tells us how often our model correctly predicts the CSAT score (1, 2, 3, 4, or 5) out of all customer interactions.

*   **Business Impact:**
    *   **High Accuracy:** A high overall accuracy means the model is generally good at predicting customer satisfaction levels. This can help Flipkart understand its customer base better and make data-driven decisions. For instance, if the model predicts a CSAT of 5 with high accuracy, it confirms areas where Flipkart excels.

    *   **Low Accuracy (as observed here):** The observed accuracies (around 37-46%) are relatively low. This indicates that the models are not very reliable at predicting the *exact* CSAT score for an individual interaction. From a business perspective, relying solely on these models for precise individual predictions would be risky. However, even with low overall accuracy, the model might still be useful for identifying *trends* or *extremes*.

    *   **Challenge with Imbalance:** Given the class imbalance (many CSAT 5s), a model could achieve a decent accuracy by simply predicting '5' most of the time. Therefore, accuracy alone is often not sufficient for imbalanced datasets, and other metrics are crucial.

### 2. Precision

*   **Indication:** Precision measures, for each predicted CSAT score (e.g., when the model *predicts* a customer will give a CSAT 1), how many of those predictions were actually correct. It answers: "Of all the customers the model predicted as CSAT 1, how many *actually* gave a CSAT 1?"

*   **Business Impact:**
    *   **High Precision for CSAT 1 (Low Satisfaction):** This is critical. If the model has high precision when predicting a CSAT 1, it means that when Flipkart identifies a customer as highly dissatisfied, that prediction is very likely to be true. This allows the business to confidently prioritize these customers for proactive outreach, service recovery, or immediate investigation into their issue.

### 3. Recall (Sensitivity)

*   **Indication:** Recall measures, for each actual CSAT score (e.g., of all customers who *actually* gave a CSAT 1), how many of them did the model correctly identify. It answers: "Of all the customers who *actually* gave a CSAT 1, how many did the model *catch*?"

*   **Business Impact:**
    *   **High Recall for CSAT 1 (Low Satisfaction):** This is arguably the most important metric for a business aiming to improve customer service. High recall for CSAT 1 means the model is effective at identifying most of the truly dissatisfied customers.     *   **High Recall for CSAT 5:** While good, it's less critical than for low CSAT scores, as missing a happy customer is less damaging than missing an unhappy one.

### 4. F1-Score

*   **Indication:** The F1-Score is the harmonic mean of Precision and Recall. It provides a single score that balances both concerns. A high F1-score indicates that the model has good values for both precision and recall, meaning it correctly identifies actual cases (good recall) and makes few false positive predictions (good precision).

*   **Business Impact:**
    *   **Balanced View for Critical Classes:** For Flipkart, the F1-Score for minority classes like CSAT 1 and 2 is particularly informative. If the F1-Score for CSAT 1 is low (e.g., ~0.28-0.03), it tells us that the model struggles to balance catching dissatisfied customers with avoiding mislabeling others as dissatisfied. A higher F1-score for CSAT 1 would indicate a more robust ability to manage crucial negative customer experiences.

    *   **Overall Model Effectiveness:** The weighted average F1-Score (around 0.44-0.51) gives a general sense of the model's performance across all classes, taking imbalance into account. The current scores suggest there's significant room for improvement in overall classification quality.

````python
# Hyperparameter Tuning - Gradient Boosting

from sklearn.model_selection import RandomizedSearchCV

# Sample for fast tuning
X_sample = X_train_balanced[:10000]
y_sample = y_train_balanced[:10000]

# Parameter grid
param_grid_gb = {
    'n_estimators': [30, 50, 100],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

# RandomizedSearchCV
random_search_gb = RandomizedSearchCV(
    GradientBoostingClassifier(random_state=42),
    param_grid_gb,
    n_iter=5,
    cv=3,
    scoring='accuracy',
    n_jobs=-1,
    random_state=42
)

random_search_gb.fit(X_sample, y_sample)

# Train best model on full data
best_gb_model = GradientBoostingClassifier(**random_search_gb.best_params_, random_state=42)
best_gb_model.fit(X_train_balanced, y_train_balanced)
y_pred_gb_tuned = best_gb_model.predict(X_test)

# Results
gb_tuned_accuracy = accuracy_score(y_test, y_pred_gb_tuned)

print("Best Parameters:", random_search_gb.best_params_)
print(f"Before Tuning: {gb_accuracy:.4f}")
print(f"After Tuning:  {gb_tuned_accuracy:.4f}")
print(f"Improvement:   {(gb_tuned_accuracy - gb_accuracy)*100:.2f}%")
````

##### Which hyperparameter optimization technique have you used and why?

I used RandomizedSearchCV for hyperparameter optimization.

Why RandomizedSearchCV?

Efficiency over GridSearchCV: Gradient Boosting, like Random Forest, has several hyperparameters that can significantly impact performance. GridSearchCV would explore every single combination, which quickly becomes computationally prohibitive, especially with cross-validation.

Effective Exploration: RandomizedSearchCV samples a fixed number of parameter settings from the specified distributions. This stochastic approach is often effective at finding good hyperparameter combinations in a much shorter time, particularly when dealing with many hyperparameters or a wide search space.

Resource Management: Training Gradient Boosting models can be intensive. By using RandomizedSearchCV, we can efficiently allocate computational resources to explore a diverse set of hyperparameters without expending excessive time on exhaustive searches.

Practicality: For complex models and datasets, RandomizedSearchCV provides a practical balance between thoroughness and computational feasibility, making it a preferred choice when time and resources are constraints.

Feature selection focused on retaining relevant and interpretable features while dropping IDs, raw timestamps, and high-cardinality nominal features like Agent_name and Customer_City to prevent overfitting and maintain interpretability.
Numerical features were scaled using StandardScaler to prepare them for distance-based and gradient-descent optimized models.

2. Exploratory Data Analysis (EDA) Insights:

CSAT Score Distribution: The majority of customers provided a CSAT score of 5 (69.4%), followed by 1 (13.1%) and 4 (13.1%), showing a bimodal distribution with fewer neutral scores (2s and 3s).

Channel Effectiveness: Outcall and Inbound channels had similar and higher average CSAT scores compared to Email, which showed the lowest satisfaction (avg CSAT ~3.89). This indicates email is a critical area for improvement.

Handling Time vs. CSAT: No clear linear relationship was observed between connected_handling_time and CSAT. Both very short and very long handling times could lead to lower satisfaction, while higher CSAT scores were observed within a moderate handling time range.

Category Impact: The 'Others' category had the lowest average CSAT (~3.43), highlighting issues with poorly defined or complex problems. 'Cancellation' and 'Product Queries' also had lower CSAT, while 'App/website' and 'Payments related' performed best.

Agent Shift & Tenure: The 'Morning' shift showed the lowest average CSAT, suggesting operational challenges during these hours. 'On Job Training' agents also had the lowest CSAT, which is expected but points to a need for better onboarding support.

Product Category: 'GiftCard' issues led to significantly low CSAT (~3.23), and physical goods like 'Furniture' and 'Mobile' also had lower satisfaction, indicating specific product-related pain points.

Item Price: Higher Item_price showed a weak negative correlation with CSAT, suggesting issues with more expensive items lead to greater dissatisfaction.
